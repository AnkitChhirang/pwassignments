```python
#Problem 1: Bank Account Create a class representing a bank account with attributes like account number, 
#account holder name, and balance. Implement methods to deposit and withdraw money from the account.

class Bank_Account():
    def __init__(self,account_number,account_holder_name,balance):
        self.account_number=account_number
        self.account_holder_name=account_holder_name
        self.balance=balance
    def deposit(self):
        deposit1=float(input('enter the amount to deposit'))
        self.balance+=deposit1
        print(f'balance after depositing entered amount is{self.balance}')
    def withdrawal(self):
        withdrawl1=float(input('enter the amount to deposit'))
        self.balance-=withdrawl1
        print(f'balance after withdrawl is{self.balance}')
```

```python
#Problem 3: Vehicle Rental Create a class representing a vehicle rental system. Implement methods to rent a 
#vehicle, return a vehicle, and display available vehicles.

class VehicleRentalSystem:
    def __init__(self):
        self.available_vehicles=set()
    def add_a_vehicle(self, vehicle):
        self.available_vehicles.add(vehicle)
        print(f'{vehicle} has been added to the list of available vehicles.')
    def display_available_vehicles(self):
        print('List of available vehicles:')
        for vehicle in self.available_vehicles:
            print(f'{vehicle.vehicle_name}-Year:{vehicle.vehicle_year_made}')
    def rent_a_vehicle(self):
        if not self.available_vehicles:
            print('No vehicles available for rent.')
            return
        print('List of available vehicles:')
        for vehicle in self.available_vehicles:
            print(f'{vehicle.vehicle_name}-Year:{vehicle.vehicle_year_made}')
        rent_choice=input('Select a vehicle from the above list: ')
        found_vehicle=None
        for vehicle in self.available_vehicles:
            if vehicle.vehicle_name==rent_choice:
                found_vehicle=vehicle
                break
        if found_vehicle:
            self.available_vehicles.remove(found_vehicle)
            print(f'{found_vehicle.vehicle_name} rented successfully.')
        else:
            print(f'{rent_choice} is not available at the moment, or recheck your input')
class Vehicle:
    def __init__(self,vehicle_name,vehicle_year_made):
        self.vehicle_name=vehicle_name
        self.vehicle_year_made=vehicle_year_made
    def __str__(self):
        return f'{self.vehicle_name}-Year:{self.vehicle_year_made}'
```

```python
#Problem 5: Product Inventory Create classes representing a product and an inventory system. Implement methods to add 
#products to the inventory, update product quantity, and display available products.

class Product:
    def __init__(self,product_id,product_name,product_quantity,price):
        self.product_id=product_id
        self.product_name=product_name
        self.product_quantity=product_quantity
        self.price=price
    def __str__(self):
        return f"{self.product_name} (ID:-{self.product_id}, Price:Rs/-{self.price}, Quantity:-{self.product_quantity})"
class Inventory:
    def __init__(self):
        self.products={}
    def add_product(self, product):
        if product.product_id not in self.products:
            self.products[product.product_id]=product
            print(f'{product.product_name} added to the inventory.')
        else:
            print(f'{product.product_name} already exists in the inventory.')
    def update_product_quantity(self,product_id,new_quantity):
        if product_id in self.products:
            self.products[product_id].quantity=new_quantity
            print(f'Quantity for {self.products[product_id].product_name} updated to {new_quantity}.')
        else:
            print(f'product with ID {product_id} not found in the inventory.')
    def display_available_products(self):
        print('available products:')
        for product in self.products.values():
            print(product)
```

```python
#Problem 6: Shape Calculation Create a class representing a shape with attributes like length, width, and height.
#Implement methods to calculate the area and perimeter of the shape.

class Shape:
    def __init__(self,length,width,height):
        self.length=length
        self.width=width
        self.height=height

class Rectangle(Shape):
    def perimeter_of_rectangle(self):
        perimeter_of_rectangle=(self.length+self.width)*2
        print(f'perimeter of this rectangle with height: {self.length}, and width: {self.width} is: {perimeter_of_rectangle}')
    def area_of_rectangle(self):
        area_of_rectangle=(self.length*self.width)
        print(f'area of this rectangle with length: {self.length}, and width: {self.width} is: {area_of_rectangle}')

class Triangle(Shape):
    def perimeter_of_triangle(self):
        perimeter_of_triangle=(self.length+self.height+self.width)
        print(f'perimeter of this triangle with height: {self.height}, length: {self.length}, and width: {self.width} is: {perimeter_of_triangle}')
    def area_of_triangle(self):
        area_of_triangle=(self.width*self.height)/2
        print(f'area of this triangle with height: {self.height}, and width: {self.width} is: {area_of_triangle}')
    
class Square(Shape):
    def area_of_square(self):
        area_of_square=(self.height*self.height)
        print(f'area of this sqaure with side: {self.height} is: {area_of_square}')
    def perimeter_of_square(self):
        perimeter_of_square=4*self.height
        print(f'perimeter of this square with side: {self.height}, is: {perimeter_of_square}')
```

```python
#Problem 7: Student Management Create a class representing a student with attributes like student ID, name, and grades. 
#Implement methods to calculate the average grade and display student details.

class Student:
    def __init__(self,student_id,name,grades):
        self.student_id=student_id
        self.name=name
        self.grades=grades
    def display_student_details(self):
        print(f'Name of the Student: {self.name}, Student_Id: {self.student_id}, Student_Grades: {self.grades}')
    def average_grades(self):
        if not self.grades:
            print("No grades available.")
            return
        sum_of_grades=sum(self.grades)
        average_grades=sum_of_grades/len(self.grades)
        print(f'average grades of {self.name} are {average_grades}')
```

```python
#Problem 9: Social Media Profile Create a class representing a social media profile with attributes 
#like username and posts. Implement methods to add posts, display posts, and search for posts by keyword.

class SocialMediaProfile:
    def __init__(self,username):
        self.username=username
        self.posts=[]
    def write_a_post(self, content):
        post = Post(content)
        self.posts.append(post)
        print('post has been added to the profile')
    def display_posts(self):
        print(f'posts of {self.username}')
        for post in self.posts:
            post.display_post_details()
    def search_post(self,keyword):
        similar_posts=[post for post in self.posts if keyword.lower() in post.content.lower()]
        print(f"posts holding '{keyword}' by {self.username}:")
        for similar_post in similar_posts:
            similar_post.display_post_details()

class Posts:
    def __init__(self, content):
        self.content = content
    def display_post_details(self):
        print(f"Content: {self.content}")
```

```python
#Problem 10: ToDo List Create a class representing a ToDo list with attributes like tasks and due dates. Implement methods to add tasks, 
#mark tasks as completed, and display pending tasks.

class ToDoList:
    def __init__(self):
        self.taskjar=[]
    def add_task(self,task,due_date):
        task_details={'task':task, 'due_date':due_date, 'status':'pending'}
        self.taskjar.append(task_details)
        print(task, 'has been added to the ToDoList')
    def mark_task_as_done(self,taskname):
        for task in self.taskjar:
            if task['task']==taskname:
                task['status']='complete'
        print(f'task "{taskname}" has been marked as completed')
        break
    def display_pending_tasks(self):
        pending_tasks = [task for task in self.taskjar if task['status'] == 'pending']
        if pending_tasks:
            print('Pending tasks:')
            for task in pending_tasks:
                print(task)
        else:
            print('There are no pending tasks.')
```
